```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
df_cal_tex_2017 = df[( (df["State"] == "California") | (df["State"] == "Texas") ) & \
   ( (df["Order Date"] >= "2017-1-1") & (df["Order Date"] <= "2018-1-1") )]
df_cal_tex_2017
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
df_agg_2017 = df[(df["Order Date"] >= "2017-1-1") & (df["Order Date"] <= "2018-1-1")]
df_agg_2017 = df_agg_2017["Sales"].agg(["sum", "mean", "std"])
df_agg_2017
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df_seg_high = df[(df["Order Date"] >= "2018-1-1") & (df["Order Date"] <= "2019-1-1")]
df_seg_high = df_seg_high.groupby("Segment")["Profit"].max()
df_seg_high
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df_top5 = df[(df["Order Date"] >= "2019-4-15") & (df["Order Date"] <= "2019-12-31")]
df_top5 = df_top5.groupby("State")["Sales"].agg("min").reset_index()
df_top5 = df_top5.sort_values("Sales", ascending=True).head(5)
df_top5
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
df_pro_sale = df[((df["Region"] == "West") | (df["Region"] == "Central")) & \
    ((df["Order Date"] >= "2019-1-1") & (df["Order Date"] <= "2019-12-31"))]
df_pro_sale = df_pro_sale.groupby("Segment").sum()
# หาสัดส่วน (Proportion) ด้วยสมการ (value / sum of value) * 100
df_pro_sale["Proportion"] = (df_pro_sale["Sales"] / df_pro_sale["Sales"].sum()) * 100
df_pro_sale["Proportion"].round(2)
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
df_top10 = df[(df["Order Date"] >= "2019-1-1") & (df["Order Date"] <= "2020-12-31")]
# 10 popular products in terms of number of orders
df_top10_1 = df_top10.groupby("Product Name")[["Quantity", "Sales"]].agg("sum").reset_index()
df_top10_1 = df_top10_1.sort_values(["Quantity", "Sales"], ascending=False).head(10)
# 10 popular products in terms of total sales
df_top10_2 = df_top10.groupby("Product Name")[["Sales", "Quantity"]].agg("sum").reset_index()
df_top10_2 = df_top10_2.sort_values(["Sales", "Quantity"], ascending=False).head(10)
print(df_top10_1)
print(df_top10_2)
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
# Segment has the highest profit in 2018
df_seg_high.plot(kind="pie", autopct="%.2f", fontsize = 15)

# top 10 popular products in terms of number of orders vs. total sales during 2019-2020
df_top10_1.plot(x = "Product Name", y = "Quantity", kind="bar")
df_top10_2.plot(x = "Product Name", y = "Sales", kind="bar")
```
